RFMÏùÑ ÌÜµÌï¥ ÏÇ¨Ïö©Ïûê ÌäπÏÑ±Î≥ÑÎ°ú Îã§Î•∏ Ï†ïÏ±ÖÏùÑ Ï†ÅÏö©ÌïòÍ≥† ÏÑúÎπÑÏä§Î•º Îçî Ïûò ÏÇ¨Ïö©ÌïòÎèÑÎ°ù Ïú†ÎèÑÌïòÎäî Ï†ÑÎûµÏùÑ ÏÑ∏Ïö∏ Ïàò ÏûàÎã§.

```python
import pandas as pd

cp_df = pd.read_csv('./datasets/customer_personality_analysis.csv', sep='\t')
cp_df
```

```python
cp_df.info()
```

```python
cp_df.duplicated().sum()
cp_df.ID.duplicated().sum()
```

```python
cp_df.isna().sum()
```

```python
cp_df[['Income']].describe().T
```

```python
cp_df['Income'].fillna(cp_df.Income.median(), inplace=True)
cp_df.isna().sum()
```

```python
cp_df.Year_Birth.min()
```

```python
cp_df = cp_df[cp_df.Year_Birth >= 1925]
cp_df
```

```python
cp_df[cp_df.Income <= 0]
```

```python
import numpy as np

Q1 = cp_df.describe().T.loc['Income','25%']
Q3 = cp_df.describe().T.loc['Income','75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'Ï†ïÏÉÅÏπò Î≤îÏúÑ: {lower_bound} ~ {upper_bound}')
```

```python
cp_df = cp_df[cp_df.Income <= upper_bound]
cp_df
```
