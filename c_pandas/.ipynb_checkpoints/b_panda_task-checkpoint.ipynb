```python
happiness_df = pd.read_csv('./datasets/happiness_report_2023.csv')
happiness_df.describe().T
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = np.percentile(happiness_df.happiness_score, 25)
happiness_Q3 = np.percentile(happiness_df.happiness_score, 75)

iqr_value = iqr(happiness_df.happiness_score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.6 * iqr_value
upper_bound = happiness_Q3 + 1.6 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
condition1 = happiness_df.happiness_score >= lower_bound
condition2 = happiness_df.happiness_score <= upper_bound
condition = condition1 & condition2

happiness_df = happiness_df[condition]
happiness_df
```

```python
hp_mean_df = happiness_df.groupby('perceptions_of_corruption').mean('happiness_score').reset_index()
hp_mean_df
```

```python
hp_min_df = happiness_df.groupby('perceptions_of_corruption').min('happiness_score').reset_index()
hp_min_df
```

```python
hp_max_df = happiness_df.groupby('perceptions_of_corruption').max('happiness_score').reset_index()
hp_max_df
```

```python
hp_std_df = happiness_df.groupby('perceptions_of_corruption')['happiness_score'].std().reset_index()
hp_std_df
```
